#### Градиентный спуск

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import derivative
plt.style.use("dark_background")
```

***Градиент*** - направление наискорейшего роста, а антиградиент - направление наискорейшего убывания.

***Градиентный спуск*** - метод нахождения локального экстремума функции с помощью движения вдоль градиента (в сторону антиградиента):

$f(x): \mathbb{R^n} \to \mathbb{R}, f(x)\to \underset {x \in \mathbb{X}}{min} \Rightarrow x_{n+1} = x_n - \gamma_n \nabla f(x_n)$, где  $\gamma_n$ - градиентный шаг градиентного спуска.

```python
def f(x):
    return x**2
xx, yy = [], []
x = 10
gamma = 0.1
for step in range(20):
    x = x - gamma*derivative(f, x, dx=1e-6)
    xx.append(x)
    yy.append(f(x))

plt.scatter(xx, yy, color='r', label='gradient descent')
xx = np.arange(-10, 10)
plt.plot(xx, f(xx), label = '$x^2$')
plt.title('Градиентный спуск')
plt.legend()
plt.show()
```
