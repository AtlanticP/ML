### Проверка стационарности и STL-декомпозиции

#### Критерий Дики-Фуллера:
$y^T = y_1, ..., y_T $

$H_0: \text{ряд нестационарен}$

$H_1: \text{ряд стационарен}$

$F_0: \text{табличное}$


```python
from pylab import rcParams
rcParams['figure.figsize'] = 18, 10
sm.tsa.seasonal_decompose(df.sales).plot()
DFC = sm.tsa.stattools.adfuller(df.sales)[1].round(4)
txt = f'Критерий Дики-Фуллера: {DFC}'
print(txt)
plt.show()
```

### Стабилизация дисперсии

Сделаем преобразование Бокса-Кокса для стабилизации дисперсии:

$
\begin{equation*}
 y_t^{'} = 
     \begin{cases}
       \textbf {ln}\; y_t,                 \;\;\;\;\;\;\;\; \lambda=0 \\
       (y_t^\lambda-1)/\lambda, \;\; \lambda \neq 0
     \end{cases}
\end{equation*}
$

```python
# inverse of Box-Cox transformation
def invboxcox(y, lmb):
    if lmb == 0:
        return(np.exp(y))
    else:
        return np.exp(np.log(lmb*y+1)/lmb)
```

```python
df['sales_box'], lmb = sts.boxcox(df.sales)
sm.tsa.seasonal_decompose(df.sales_box).plot()
DFC = sm.tsa.stattools.adfuller(df.sales_box)[1].round(4)
txt = f'Критерий Дики-Фуллера: {DFC}'
print(txt)
plt.show()
```

```python
sm.graphics.tsa.plot_acf(df['sales'].values,lags=48)
plt.show()
```

Согласно критерию Дики-Фуллера ряд стационарный (гипотеза о нестационарности отвергается). Однако стационарным его считать нельзя т.к. виден тренд и видна сезонность.

### Диффиренцирование ряда 
переход к попарным разностям соседних значений:
$$y_t{'} = y_t - y_{t-1} \notag$$
- стабилизация среднего значения, избавление от тренда

#### Сезонное дифференцирования
переход к попарным разностям значений в соседних сезонах: 
$$y_t{'} = y_t - y_{t-s} \notag$$
- убирает сезонность

```python
# убираем сезонность
df['sales_box_diff'] = df.sales_box - df.sales_box.shift(12)
rcParams['figure.figsize'] = 18, 10
sm.tsa.seasonal_decompose(df.sales_box_diff[12:]).plot()
DFC = sm.tsa.stattools.adfuller(df.sales_box_diff[12:])[1].round(4)
txt = f'Критерий Дики-Фуллера: {DFC}'
print(txt)
plt.show()
```

```python
rcParams['figure.figsize'] = 18, 3
sm.graphics.tsa.plot_acf(df.sales_box_diff.values[13:],lags=48)
plt.show()
```

Сезонность была убрана. Однако согласно критерию Дики-Фуллера отвергнуть нулевую гипотезу о нестационарности ряда нельзя. Применяю обычное дифференцирование.

```python
rcParams['figure.figsize'] = 18, 10
df['sales_box_diff2'] = df.sales_box_diff - df.sales_box_diff.shift(1)
sm.tsa.seasonal_decompose(df.sales_box_diff2[13:]).plot()
DFC = sm.tsa.stattools.adfuller(df.sales_box_diff2[13:])[1].round(6)
txt = f'Критерий Дики-Фуллера: {DFC}'
print(txt)
plt.show()
```

Гипотеза нестационарности отвергается, и визуально ряд выглядит лучше - тренда больше нет.

### Подбор модели

Посмотрим на ACF и PACF полученного ряда:

```python
rcParams['figure.figsize'] = 18, 4
sm.graphics.tsa.plot_acf(df.sales_box_diff2[13:].values,lags=48)
sm.graphics.tsa.plot_pacf(df.sales_box_diff2[13:].values,lags=48)
plt.show()
```

Начальные приближения: Q=1, q =2, P=1, p=4;
Q*S - максимальный сезонный лаг значимо отличающийся от нуля на АС равен 12;
q - максимальный лаг значимо отличающийся от нуля на АС равен 2;
P*S - максимальный сезонный лаг значимо отличающийся от нуля на PАС равен 12;
q - максимальный лаг значимо отличающийся от нуля на PАС равен 4.

```python
ps = range(0, 5)
d = 1
qs = range(0, 3)
Ps = range(0, 2)
D = 1
Qs = range(0, 2)
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
result = []
best_aic = float('inf')
warnings.filterwarnings('ignore')

for param in parameters_list:

    try:
        model = sm.tsa.statespace.SARIMAX(
            df.sales_box, 
            order=(param[0], d, param[1]),
            seasonal_order=(param[2], D, param[3], 12)
        ).fit(disp=-1)
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    
    if aic < best_aic:
        best_model = model
        best_aic = aic
    
    result.append([param, model.aic])

warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(result)
result_table.columns = ['parameters', 'aic']
result_table.sort_values(by='aic', ascending=True).round(2).head()
```

#### Критерий Льюнга-Бокса
$H_0:$  данные являются случайными (белый шум)

$H_1:$ данные не являются случайными

Лучшая модель с минимальным значением информационного критерия Акаике:

```python
best_model.summary()
```

Q-тест Льюнга - Бокса - (0.01, 0.93) - данные являются случайными (белый шум, неавтокоррелированы).

#### Визуальный анализ остатков лучшей модели

```python
# Повторим Q-текст Льюнга-Бокса
q_test = sm.tsa.stattools.acf(best_model.resid[13:], qstat=True, fft=False, nlags=40)
# print(f'Q-stat:', q_test[0])
df_Q = np.vstack([q_test[1], q_test[2]]).T
df_Q = pd.DataFrame(
    df_Q, columns=['qstat', 'pvalues'], 
    index=range(1, df_Q.shape[0]+1)
    )
df_Q
```

P уровни значимости свидетельствуют о том, что данные корреляции есть результат случайности. Нулевая гипотеза об неавтокоррелированности (независимости) остатков не отвергается.

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(18, 6))
best_model.resid[13:].plot(ax=axes[0])
sm.graphics.tsa.plot_acf(best_model.resid[13:].values, lags=48, ax=axes[1])
plt.show()

DFC = sm.tsa.stattools.adfuller(best_model.resid[13:])[1]
txt = f'Критерий Дики-Фуллера: {DFC}'
print(txt)
SC = sts.ttest_1samp(best_model.resid[13:], 0)[1]
txt = f'Критерий Стьюдента: {SC.round(3)}'
print(txt)
```

Остатки несмещены (кр. Стьюдента), стационарны (кр. Д-Ф, визуально), неавткоррелированы (Льюнга-Бокс и коррелограмма).

```python
rcParams['figure.figsize'] = (13, 3)
df['model'] = invboxcox(best_model.fittedvalues, lmb)
df.sales.plot()
df.model[13:].plot(color='r', label='best_model')
plt.ylabel('Wine sales')
plt.legend()
plt.show()
```

### Прогноз

```python
df2 = df[['sales']].copy()
date_list = [datetime.datetime.strptime("1994-09-01", "%Y-%m-%d") 
             + relativedelta(months=x) for x in range(0,36)]
future = pd.DataFrame(index=date_list, columns=df.columns)
df2 = pd.concat([df2, future])

df2['forecast'] = invboxcox(best_model.predict(start=176, end=211), lmb)
df2.sales.plot()
df2.forecast.plot(color='r')
plt.show()
```

В предсказании также видна сезонность ряда и в нем нет тренда. Поскольку тренд на обучении имел сложную структуру и в момент, когда необходимо было сделать предсказание модель не смогла определить тренд (направление тренда).

Исследуем модель на предмет того, какую часть наблюдений она описывает.

***Коэффициент детерминации:***

$R^2 = 1 - \frac{\sum\limits_{i=1}^{n}(y_i-f(x_i))^2}
    {\sum\limits_{i=1}^{n}(y_i-\overline{y})^2} =
     1 - \frac{\sum\limits_{i=1}^{n}\varepsilon_i^2}
    {\sum\limits_{i=1}^{n}(y_i-\overline{y})^2}$

```python
round(r2_score(df.sales[13:], invboxcox(best_model.fittedvalues[13:], lmb)), 2)
```

81% дисперсии ряда объясняется нашей моделью.
